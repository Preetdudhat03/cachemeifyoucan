# OptionSight AI - Exploratory Data Analysis

This notebook demonstrates how to load and analyze options market data natively using the `src` modules.

```python
import sys
import os
import pandas as pd

# Add root project to python path
sys.path.append(os.path.abspath('../'))

from src.data_loader import load_options_data
from src.feature_engineering import run_feature_engineering
from src.sentiment_analysis import analyze_sentiment
from src.anomaly_detection import detect_anomalies
from src.insights_engine import generate_insights
```

```python
# Load and inspect basic data
df = load_options_data('../data/options_data.csv')
df.head()
```

```python
# Feature Engineering Pipeline
df_features = run_feature_engineering(df)
df_features[['strike', 'spot_close', 'PCR_OI', 'call_money_flow', 'put_money_flow']].head()
```

```python
# AI Anomalies & Sentiment Insights
sentiment = analyze_sentiment(df_features)
print("Market Sentiment:", sentiment)

df_anomalies = detect_anomalies(df_features)
insights = generate_insights(df_anomalies, sentiment)
print("\n--- AI Market Insights ---")
for i in insights:
    print("-", i)
```
